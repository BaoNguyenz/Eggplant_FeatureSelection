```python
import pandas as pd

df_dense = pd.read_csv(r"E:\PROJECTWORSHOP\Eggplant Leaf Disease Detection Dataset\code_file\densenet_feature_extraction\output\densenet121_features_notest.csv")
df_basic = pd.read_csv(r"E:\PROJECTWORSHOP\Eggplant Leaf Disease Detection Dataset\code_file\A_feature_selection\csv_data\basic_ad_features_all_disease.csv")
df_basic_ad = pd.read_csv(r"E:\PROJECTWORSHOP\Eggplant Leaf Disease Detection Dataset\code_file\A_feature_selection\csv_data\basic_ad_features_all_disease.csv")
df_gist = pd.read_csv(r"E:\PROJECTWORSHOP\Eggplant Leaf Disease Detection Dataset\code_file\A_feature_selection\csv_data\gist_features_all_diseases.csv")
df_glcm = pd.read_csv(r"E:\PROJECTWORSHOP\Eggplant Leaf Disease Detection Dataset\code_file\A_feature_selection\csv_data\glcm_features_all_diseases.csv")
df_hist = pd.read_csv(r"E:\PROJECTWORSHOP\Eggplant Leaf Disease Detection Dataset\code_file\A_feature_selection\csv_data\hist_features_all_disease.csv")
df_lbp = pd.read_csv(r"E:\PROJECTWORSHOP\Eggplant Leaf Disease Detection Dataset\code_file\A_feature_selection\csv_data\lbp_features_all_diseases.csv")
```

```python
df_dense.shape, df_basic_ad.shape, df_basic.shape, df_gist.shape, df_glcm.shape, df_hist.shape, df_lbp.shape
```

```python
# remove cột đầu tiên
df_dense = df_dense.iloc[:, 1:]
df_basic = df_basic.iloc[:, 1:]
df_basic_ad = df_basic_ad.iloc[:, 1:]
df_gist = df_gist.iloc[:, 1:]
df_glcm = df_glcm.iloc[:, 1:]
df_hist = df_hist.iloc[:, 1:]
df_lbp = df_lbp.iloc[:, 1:]

# ghép với densenet
dense_basic = pd.concat([df_dense, df_basic], axis=1)
dense_basic_ad = pd.concat([df_dense, df_basic_ad], axis=1)
dense_gist = pd.concat([df_dense, df_gist], axis=1)
dense_glcm = pd.concat([df_dense, df_glcm], axis=1)
dense_hist = pd.concat([df_dense, df_hist], axis=1)
dense_lbp = pd.concat([df_dense, df_lbp], axis=1)

# lưu file
dense_basic.to_csv(r"E:\PROJECTWORSHOP\Eggplant Leaf Disease Detection Dataset\code_file\A_feature_selection\csv_data\data_concat\dense_basic.csv", index=False)
dense_basic_ad.to_csv(r"E:\PROJECTWORSHOP\Eggplant Leaf Disease Detection Dataset\code_file\A_feature_selection\csv_data\data_concat\dense_basic_ad.csv", index=False)
dense_gist.to_csv(r"E:\PROJECTWORSHOP\Eggplant Leaf Disease Detection Dataset\code_file\A_feature_selection\csv_data\data_concat\dense_gist.csv", index=False)
dense_glcm.to_csv(r"E:\PROJECTWORSHOP\Eggplant Leaf Disease Detection Dataset\code_file\A_feature_selection\csv_data\data_concat\dense_glcm.csv", index=False)
dense_hist.to_csv(r"E:\PROJECTWORSHOP\Eggplant Leaf Disease Detection Dataset\code_file\A_feature_selection\csv_data\data_concat\dense_hist.csv", index=False)
dense_lbp.to_csv(r"E:\PROJECTWORSHOP\Eggplant Leaf Disease Detection Dataset\code_file\A_feature_selection\csv_data\data_concat\dense_lbp.csv", index=False)

# # kiểm tra shape
print("dense_basic:", dense_basic.shape)
print("dense_basic_ad:", dense_basic.shape)
print("dense_gist:", dense_gist.shape)
print("dense_glcm:", dense_glcm.shape)
print("dense_hist:", dense_hist.shape)
print("dense_lbp:", dense_lbp.shape)
```
